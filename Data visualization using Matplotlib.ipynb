```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100).reshape(-1,1)
```

```python
y = np.arange(0,100).reshape(-1,1)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('scatter plot')
plt.show()
```

```python
x = x*x
```

```python
plt.plot(x, y, 'c--', marker='o', linestyle='dashed',  linewidth=3, markersize=5)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'b--')
plt.show()
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.plot(x,y_sin,'r--')
plt.title('sin wave')

plt.subplot(1,2,2)
plt.plot(x,y_cos,'b--')
plt.title('cosine wave')

plt.show()
```
